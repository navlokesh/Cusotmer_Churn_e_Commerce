## 1) Introduction:

### Problem Statement:

#### The data set belongs to a leading online E-Commerce company. An online retail (E commerce) company wants to know the customers who are going to churn, so accordingly they can approach customer to offer some promos.


### Need for Solving it:

Customer retention: 
It is one of the very important aspect for business. 
Once a customer goes to an e-Commerce website and order something there is a possibility that the customer would come back and buy more things (If they are happy with the experience). 
Customer retention helps in generating higher customer lifetime value and hence the revenue.
It is good to have new customers, but existing customers bring more revenue than the new ones.

There are many ways to achieve customer retention but the most commonly used model is- Churn Model.
Churn Model helps identifying customers who are most likely to switch to different e-Commerce website. Once identified the company can take actions in order to keep its existing customers. Now the question is, how does Churn model identify these customers? The model can be used to calculate the churn rate and depending on the nature of business, different metrics can be used. Few common metrics are -
•	Number of customers lost
•	Percent of customers lost

### Business/Social Opportunity:

The Churn Model provides opportunities to the businesses in many ways. Few advantages of implementing Churn Model in e-Commerce are -
•	Churn rate can help identifying churn customers and accordingly businesses can run retention campaigns.
•	It will help in the keeping the revenue flowing.
•	Churn Model can help business to maintain customer lifetime value.
•	It helps businesses to track the progress.
•	The inputs received from Churn Model can be very helpful for BI activities.

### Outlier treatment:


```python
plt.figure(figsize=[20,10])
df.boxplot()
plt.xticks(rotation=90)
```

```python
### Treating outliers with the IQR range for the interger variables
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lower,upper=remove_outlier(df['Tenure'])
df['Tenure']=np.where(df['Tenure']>upper,upper,df['Tenure'])
df['Tenure']=np.where(df['Tenure']<lower,lower,df['Tenure'])

lower,upper=remove_outlier(df['WarehouseToHome'])
df['WarehouseToHome']=np.where(df['WarehouseToHome']>upper,upper,df['WarehouseToHome'])
df['WarehouseToHome']=np.where(df['WarehouseToHome']<lower,lower,df['WarehouseToHome'])

lower,upper=remove_outlier(df['HourSpendOnApp'])
df['HourSpendOnApp']=np.where(df['HourSpendOnApp']>upper,upper,df['HourSpendOnApp'])
df['HourSpendOnApp']=np.where(df['HourSpendOnApp']<lower,lower,df['HourSpendOnApp'])

lower,upper=remove_outlier(df['NumberOfDeviceRegistered'])
df['NumberOfDeviceRegistered']=np.where(df['NumberOfDeviceRegistered']>upper,upper,df['NumberOfDeviceRegistered'])
df['NumberOfDeviceRegistered']=np.where(df['NumberOfDeviceRegistered']<lower,lower,df['NumberOfDeviceRegistered'])

lower,upper=remove_outlier(df['NumberOfAddress'])
df['NumberOfAddress']=np.where(df['NumberOfAddress']>upper,upper,df['NumberOfAddress'])
df['NumberOfAddress']=np.where(df['NumberOfAddress']<lower,lower,df['NumberOfAddress'])

lower,upper=remove_outlier(df['OrderAmountHikeFromlastYear'])
df['OrderAmountHikeFromlastYear']=np.where(df['OrderAmountHikeFromlastYear']>upper,upper,df['OrderAmountHikeFromlastYear'])
df['OrderAmountHikeFromlastYear']=np.where(df['OrderAmountHikeFromlastYear']<lower,lower,df['OrderAmountHikeFromlastYear'])

lower,upper=remove_outlier(df['CouponUsed'])
df['CouponUsed']=np.where(df['CouponUsed']>upper,upper,df['CouponUsed'])
df['CouponUsed']=np.where(df['CouponUsed']<lower,lower,df['CouponUsed'])


lower,upper=remove_outlier(df['OrderCount'])
df['OrderCount']=np.where(df['OrderCount']>upper,upper,df['OrderCount'])
df['OrderCount']=np.where(df['OrderCount']<lower,lower,df['OrderCount'])


lower,upper=remove_outlier(df['DaySinceLastOrder'])
df['DaySinceLastOrder']=np.where(df['DaySinceLastOrder']>upper,upper,df['DaySinceLastOrder'])
df['DaySinceLastOrder']=np.where(df['DaySinceLastOrder']<lower,lower,df['DaySinceLastOrder'])


lower,upper=remove_outlier(df['CashbackAmount'])
df['CashbackAmount']=np.where(df['CashbackAmount']>upper,upper,df['CashbackAmount'])
df['CashbackAmount']=np.where(df['CashbackAmount']<lower,lower,df['CashbackAmount'])
```

## 4. Model building

```python
df1 = df.drop('Churn', axis = 1, inplace=False)
```

```python
for feature in df1:
    if df1[feature].dtype == 'object':
        df1[feature] = pd.Categorical(df1[feature]).codes
```

```python
x = df1.copy()
y = df.Churn
```
